# Group by and Aggregating

```python
df3 = pd.read_csv("/Users/dibaspaudel/Desktop/Python/PandasYouTubeSeries/Flavors.csv")
```

```python
df3
```

```python
group_by_frame = df3.groupby("Base Flavor")
```

```python
group_by_frame.mean()
```

```python
df3.groupby("Base Flavor").mean()
```

```python
df3.groupby("Base Flavor").count()
```

```python
df3.groupby("Base Flavor").max()
```

```python
#df3[df3.Flavor.str.contains('Chocolate')]

df3[df3.Flavor == 'Chocolate']
```

```python
df3.groupby("Base Flavor").agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

```python
df3.groupby(['Base Flavor', 'Liked']).mean()
```

```python
df3.groupby('Base Flavor').describe()
```

# Merge, Join and Concatenate

```python
# inner join - (A n B)
# outer join - (A u B)
# left outer join - (A + (A n B))
# right outer join - (B + (B n A))
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv("/Users/dibaspaudel/Desktop/Python/PandasYouTubeSeries/LOTR.csv")
df1
```

```python
df2 = pd.read_csv("/Users/dibaspaudel/Desktop/Python/PandasYouTubeSeries/LOTR 2.csv")
df2
```

```python
# Merge

df1.merge(df2)
```

```python
    df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
    df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```
